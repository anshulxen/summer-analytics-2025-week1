### 1. What is name of car that has the highest horsepower?  

```python
# Reload from scratch
df = pd.read_csv('/kaggle/input/cars-sa-anshul/Cars.csv')

# Clean horsepower
df['horsepower'].replace('?', np.nan, inplace=True)
df['horsepower'] = df['horsepower'].astype(float)
df['horsepower'].fillna(df['horsepower'].median(), inplace=True)

# Add hp_per_weight 

df['hp_per_weight'] = df['horsepower'] / df['weight']

# Find the car with max horsepower
max_hp_car = df[df['horsepower'] == df['horsepower'].max()]
print(" Car with highest horsepower:", max_hp_car['name'].values[0])
```

### 2. How many cars have mpg  35?

```python
high_mpg_count = df[df['mpg'] >= 35].shape[0]
print(" Number of cars with mpg  35:", high_mpg_count)
```

### 3. What is the most common origin for cars with horsepower > 100 and weight < 3000?

```python
filtered_df = df[(df['horsepower'] > 100) & (df['weight'] < 3000)]
common_origin = filtered_df['origin'].value_counts().idxmax()
print(" Most common origin:", common_origin)
```

### 4. What is the mean acceleration of cars from Japan? (rounded to 2 decimals)

```python
mean_acc_japan = df[df['origin'] == 'japan']['acceleration'].mean().round(2)
print(" Mean acceleration (Japan):", mean_acc_japan)
```

### 5. Which year had the highest average mpg?

```python
best_mpg_year = df.groupby('model_year')['mpg'].mean().idxmax()
print("Year with highest avg mpg:", best_mpg_year)
```

```python
# Step 1: Calculate median mpg
median_mpg = df['mpg'].median()

# Step 2: Filter cars with above-median mpg
above_median_mpg = df[df['mpg'] > median_mpg]

# Step 3: Compute horsepower-to-weight ratio
above_median_mpg['hp_to_weight'] = above_median_mpg['horsepower'] / above_median_mpg['weight']

# Step 4: Find max ratio
max_ratio = above_median_mpg['hp_to_weight'].max()

# Step 5: Get the car(s) with max ratio
best_perf_cars = above_median_mpg[above_median_mpg['hp_to_weight'] == max_ratio]

best_perf_cars
```

### Design a multi-line plot using Matplotlib or Seaborn that shows the evolution of average mpg over the years, separately for each origin

```python
# Grouping data by 'model_year' and 'origin' to calculate average MPG
# This helps visualize how fuel efficiency evolved in different regions over time

avg_mpg_by_year_origin = df.groupby(['model_year', 'origin'])['mpg'].mean().reset_index()

# Setting the visual theme
plt.figure(figsize=(12, 6))
sns.set_style("whitegrid")

# Plotting lineplot with Seaborn to track average MPG across years, grouped by origin
sns.lineplot(
    data=avg_mpg_by_year_origin,
    x='model_year',
    y='mpg',
    hue='origin',
    marker='o',
    palette='Set2'
)

# Title & axis labels with professional formatting
plt.title('Trend of Average MPG Over the Years by Origin', fontsize=15)
plt.xlabel('Model Year', fontsize=12)
plt.ylabel('Average MPG', fontsize=12)
plt.legend(title='Car Origin')
plt.xticks(avg_mpg_by_year_origin['model_year'].unique())  # Ensures all years are shown
plt.tight_layout()
plt.show()
```

### We define a consistent car model as one that was produced over multiple years and had very low variation in mpg across those years (standard deviation < 1.0).

Tasks:

Identify car names that appear in more than one model_year.

For each such name, compute the standard deviation of mpg across years.

Return the car(s) with the lowest variation in mpg, among those with at least 2 appearances and std(mpg) < 1.0.

Report the model name(s), number of appearances, and the average mpg.

Bonus:
Sort the result by number of appearances (descending), then mpg (descending).

```python
# Reset index to get 'name' back as a column
df_reset = df.reset_index()

# Group by 'name' and filter those that appear in more than one year
multi_year_models = df_reset.groupby('name').filter(lambda x: x['model_year'].nunique() > 1)

# Group again and compute stats
consistency_df = multi_year_models.groupby('name').agg(
    appearances=('model_year', 'nunique'),
    mpg_std=('mpg', 'std'),
    avg_mpg=('mpg', 'mean')
).reset_index()

# Filter for consistent models
consistent_cars = consistency_df[consistency_df['mpg_std'] < 1.0]

# Bonus: Sort as required
consistent_cars_sorted = consistent_cars.sort_values(
    by=['appearances', 'avg_mpg'],
    ascending=[False, False]
)

consistent_cars_sorted
```
